CSV to DataFrame

```python
mw_raw_data = pd.read_csv('/Users/burhanuddinshakir/Desktop/Burhan/Work/Profcal/M&W Generic Data Work/McCance___Widdowson_s_Composition_of_Foods_Integrated_Dataset/Factors-Table 1.csv')
mw_raw_data = mw_raw_data.drop(mw_raw_data.index[[0,1]])
mw_raw_data
```

```python
food_names = list(mw_raw_data['Food Name'])
food_names
```

```python
ingredient_list = []
for food_item in food_names:
    ingredient = food_item.split(',')[0]
    ingredient_list.append(ingredient)
ingredient_list
```

Removing duplicates

```python
ingredient_list = list(set(ingredient_list))
ingredient_list
```

```python
main_ingredients_df = pd.DataFrame({'ingredient_name':ingredient_list})
main_ingredients_df
```

```python
main_ingredients_df.index.name = 'id'
main_ingredients_df
```

```python
id = main_ingredients_df[main_ingredients_df['ingredient_name'].str.contains("Ackee")].index[0]
print(id)
```

```python
id = main_ingredients_df.index[main_ingredients_df['ingredient_name'].str.contains("Baking fat and margarine \(75-90% fat\)")].values[0]
print (id)
```

```python
product_names = []
for food_item in food_names:
    ingredient = food_item.split(',')[0]
    product_names.append(ingredient)
product_names
```

```python
main_ingredient_id_list = []

for product in product_names:
    
    product = product.replace('(','\(')
    product = product.replace(')','\)')
    id = main_ingredients_df.index[main_ingredients_df['ingredient_name'].str.contains(product)].values[0]
    
    main_ingredient_id_list.append(id)
    
main_ingredient_id_list
```

```python
mw_raw_data['main_ingredient_code'] = main_ingredient_id_list
mw_raw_data
```
